## Leitura dos dados

```python
caminho_arquivo_excel = '/content/drive/MyDrive/FIAP - Pós tech/Exportacao_BD.xlsx'
```

Leitura da primeira aba "Consolidado Quantidade"

```python
df_cons_quantidade = pd.read_excel(caminho_arquivo_excel, sheet_name='Consolidado Quantidade')
```

Correção dos nomes dos países

```python
df_cons_quantidade['País'] = df_cons_quantidade['País'].replace(dicionario_paises)
```

```python
df_cons_quantidade = df_cons_quantidade.drop_duplicates(subset='País')
```

## Primeiros cálculos

Cálculo das estatísticas de quantidade

```python
df_cons_quantidade.loc[:, 'Quantidade_Total_(kg)'] = df_cons_quantidade.iloc[:, 2:].sum(axis=1)
df_cons_quantidade.loc[:, 'Quantidade_Media_(kg)'] = df_cons_quantidade.iloc[:, 2:].mean(axis=1)
df_cons_quantidade.loc[:, 'Quantidade_DesvPad_(kg)'] = df_cons_quantidade.iloc[:, 2:].std(axis=1)
```

Cálculo das estatísticas de vendas

```python
df_cons_vendas.loc[:, 'Vendas_Total_(US$)'] = df_cons_vendas.iloc[:, 2:].sum(axis=1)
df_cons_vendas.loc[:, 'Vendas_Media_(US$)'] = df_cons_vendas.iloc[:, 2:].mean(axis=1)
df_cons_vendas.loc[:, 'Vendas_DesvPad_(US$)'] = df_cons_vendas.iloc[:, 2:].std(axis=1)
```

# Análise 5: Projeções

Objetivo: Analisar os países com um alto potencial de crescimento com base em tendências passadas e projeções futuras.

Selecionar apenas as colunas de interesse

```python
columns_of_interest = ['País'] + [col for col in df_cons_quantidade.columns if isinstance(col, int)]
df_selected = df_cons_quantidade[columns_of_interest]
```

Calcular a taxa média de crescimento para cada país durante o período histórico

```python
df_selected['Taxa_Crescimento'] = df_selected.iloc[:, 1:].pct_change(axis=1).mean(axis=1) * 100
```

Remover a coluna 'Taxa_Crescimento' do dataframe selecionado

```python
df_selected = df_selected.drop('Taxa_Crescimento', axis=1)
```

Ordenar os países pela taxa de crescimento em ordem decrescente

```python
df_sorted = df_selected.sort_values(by=df_selected.columns[-1], ascending=False)
```

Calcular a projeção de crescimento para os próximos 5 anos

```python
start_year = int(df_selected.columns[-1])
end_year = start_year + 5
df_sorted['Projeção_Crescimento'] = df_sorted.iloc[:, -1] * ((1 + df_sorted.iloc[:, -1] / 100) ** (end_year - start_year))
```

Exibir os países com maior taxa de crescimento e projeção de crescimento

# Análise 6: Principais desafios e obstáculos

Objetivo: Identificar os principais desafios e obstáculos que afetam as exportações para cada país.

*    ***Barreiras tarifárias e não tarifárias:*** Tarifas de importação elevadas, cotas de importação, restrições alfandegárias, regulamentações sanitárias, normas técnicas e outras barreiras comerciais podem dificultar o acesso aos mercados internacionais.

*    ***Instabilidade política:*** A instabilidade política, conflitos internos, mudanças frequentes de governo e a falta de cooperação internacional podem criar incertezas e afetar negativamente as exportações.

*    ***Flutuações cambiais:*** Oscilações abruptas nas taxas de câmbio podem impactar a competitividade dos produtos exportados, tornando-os mais caros ou menos atrativos em relação aos produtos de outros países.

*    ***Infraestrutura deficiente:*** A falta de infraestrutura adequada, como portos, estradas, sistemas de transporte e logística, pode aumentar os custos de exportação e afetar a eficiência do comércio internacional.

*    ***Barreiras culturais e linguísticas:*** Diferenças culturais, práticas comerciais específicas e a barreira da língua podem dificultar o estabelecimento de relacionamentos comerciais e a penetração em determinados mercados.

*    ***Concorrência global:*** A competição acirrada de outros países e empresas estrangeiras pode tornar difícil a entrada em certos mercados, especialmente quando já existem produtos similares estabelecidos.

*    ***Mudanças nas políticas comerciais:*** Alterações nas políticas comerciais, acordos bilaterais ou multilaterais, restrições comerciais impostas por outros países e alterações nas preferências tarifárias podem afetar o acesso e as condições de mercado para as exportações.

*    ***Complexidade burocrática:*** Processos burocráticos complexos, exigências de documentação extensas e demoradas podem aumentar os custos e o tempo necessários para exportar produtos.
